## Step 2: Clean your data.

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, fill in any missing values. 

```python
def clean(x):
    good_char = []
    for y in x:
        if y not in ('x','.'):
            good_char.append(y)
            
    clean_col = "".join(good_char)
    return clean_col

clean_columns = []

for values in billboard.columns.values:
    new_col = clean(values)
    clean_columns.append(new_col)
    
print clean_columns

billboard.columns = clean_columns
```

```python
#making a chart that excludes weeks with all NaN values

def nanweekremover(x):
    for column in x:
        if x[column].count() == 0:
            del x[column]
    return x

newbillboard = nanweekremover(billboard)
newbillboard.head()
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
meltboard = pd.melt(billboard, id_vars = ['year', 'artistinverted', 
                              'track', 'time', 'genre', 'dateentered', 
                              'datepeaked'],
        var_name = 'Week', value_name = 'Ranking')
meltboard.head()
```

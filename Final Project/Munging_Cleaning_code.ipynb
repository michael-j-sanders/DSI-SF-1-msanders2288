```python
#function to replace NaNs with a substituted string of choice.
def denaner(x, y):
    for i in range(len(fighters)):
        if pd.isnull(fighters[x][i]) == True:
            fighters[x][i] = y
        else: 
            pass
```

```python
import datetime

total_times = []
for x in new_time:
    times  = datetime.datetime.strptime(x, '%M:%S')
    mins = times.minute
    secs = times.second
    secs = secs/60.
    mins = mins+secs
    total_times.append(mins)
```

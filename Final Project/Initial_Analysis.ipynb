```python
import pandas as pd
import numpy as np
```

```python
fights = pd.read_csv('Data Sets/Merged_Fights_Fighters.csv', parse_dates = ['f1birth_date', 'f2birth_date', 'event_date'])
```

```python
fights.drop(['Unnamed: 0', 'f1fid', 'f2fid', 'f1nick', 'f2nick'], axis = 1, inplace= True)
```

```python
fights.isnull().sum()
```

```python
def denaner(x, y):
    for i in range(len(fights)):
        if pd.isnull(fights[x][i]) == True:
            fights[x][i] = y
        else: 
            pass
```

```python
# Taking out the NaNs in various areas
denaner('f1association', 'Unassociated')
denaner('f2association', 'Unassociated')
denaner('time', '10:01')
denaner('f2birth_date', 'Date Unknown')
denaner('f1birth_date', 'Date Unknown')
```

```python
import datetime

total_times = []
for x in fights['time']:
    times  = datetime.datetime.strptime(x, f'%M:%S')
    total_times.append(times)
```

```python
fights['time'] = total_times
```

```python
new_time = []
for x in fights['time']:
    if x =='36:00:00':
        fixed_time = '36:00'
    elif x == '33:00:00':
        fixed_time = '33:00'
    elif x == '30:00:00':
        fixed_time = '30:00'
    else:
        fixed_time = x
    new_time.append(fixed_time)
new_time(fixed_time)
    
```
